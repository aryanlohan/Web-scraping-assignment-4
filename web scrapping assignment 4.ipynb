# Answer 2. Scrape the details team Indiaâ€™s international fixtures from bcci.tv. Url = https://www.bcci.tv/.

```python
browser.get('https://www.bcci.tv/')
browser.maximize_window()
time.sleep(5)
```

```python
try:
    accept_button = browser.find_element_by_xpath("//button[@class='cookie__accept btn btn-primary']")
    accept_button.click()
except:
    pass
```

```python
international_button = browser.find_element_by_xpath("//a[@href='https://www.bcci.tv/international/fixtures']")
international_button.click()
```

```python
time.sleep(5)
```

```python
fixture_cards = browser.find_elements_by_xpath("//div[@class='fixture-card-top']")
fixture = []
for i in fixture_cards:
    fixture.append(i.text.split('\n'))
fixture
```

```python
series = []
date = []
matchtime = []

for i in fixture:
    series.append(i[0])
    date.append(i[1])
    matchtime.append(i[2])
```

```python
matches = browser.find_elements_by_xpath("//div[@class='fix-place ng-binding ng-scope']")
match=[]
for i in matches:
    match.append(i.text.split('\n'))
```

```python
match
```

```python
title=[]
location=[]
for i in match:
    if str(i).find('-')!=-1:
        title.append(str(i).split('-')[0])
        location.append(str(i).split('-')[1])
    else:
        title.append('-')
        location.append(i)
```

```python
for j in range(0,len(title)):
    title[j]=title[j].replace('[','')
```

```python
bcci = pd.DataFrame()
bcci['Match Title'] = title
bcci['Series'] = series
bcci['Place'] = location
bcci['Date'] = date
bcci['Time'] = matchtime
bcci
```

```python
explore = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details").click()

trend_url = driver.find_element_by_xpath("/html/body/div[1]/header/div[1]/div[2]/nav/ul/li[4]/details/div[1]/ul[1]/li[3]/a")
urls = trend_url.get_attribute("href")
driver.get(urls)
```

```python
URLs = []
repository_title = []
Description = []
Contributors = []
Language = []
lang = []

repository = driver.find_elements_by_xpath("//h1[@class = 'h3 lh-condensed']//a")
for i in repository:
    URLs.append(i.get_attribute("href"))

title = driver.find_elements_by_xpath("//h1[@class = 'h3 lh-condensed']")
for i in title:
    repository_title.append(i.text)

for i in URLs:
    driver.get(i)
    time.sleep(3)
    
    try:
        desc = driver.find_element_by_xpath("//p[@class='f4 mt-3']")
        Description.append(desc.text)
    except NoSuchElementException:
        Description.append('-')
        

    try:
        contributor = driver.find_element_by_xpath("//*[contains(text(),'    Contributors ')]")
        Contributors.append(contributor.text.replace('Contributors',''))
    except NoSuchElementException:
        Contributors.append('-')

    try:
        for i in driver.find_elements_by_xpath("//span[@class='color-text-primary text-bold mr-1']"):
            lang.append(i.text)
        Language.append(lang)
    except NoSuchElementException:
        Language.append('-')

Github=pd.DataFrame({})
Github['Repository title'] = repository_title
Github['Repository description'] = Description
Github['Contributors count'] = Contributors
Github['Language used'] = Language
Github
```
